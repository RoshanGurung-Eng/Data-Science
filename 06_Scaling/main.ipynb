```python
# T-Test Example - Sales Data
from scipy import stats
import numpy as np 
```

```python
#Group A: Sales without a discount (per day data.)
grp_a = [200, 210, 190, 205, 195]
#Group B: sales with a discount
grp_b = [250, 240, 260, 230, 255]

t_stat, p_val = stats.ttest_ind(grp_a, grp_b)

print(f"P-Value: {p_val:.4f}")

if p_val < 0.05:
    print("The discount significally increased sales!")
else:
    print("The difference might just be luck.")
```

```python
#T-Test Example Programs - Website Performance

#Group c: Page load time (secs) with Old Server
grp_c = [3.2, 3.5, 2.9, 3.1, 3.4]
#Group d: Page load time (secs) with New Server
grp_d = [2.1, 2.3, 2.0, 1.9, 2.2]

t_stats, p_vals = stats.ttest_ind(grp_c, grp_d)

print(f"P-Value: {p_vals:.4f}")

if p_vals < 0.05:
    print("The new server significantly improved speed")
else:
    print("The speed difference in not statistically significant.")
```
