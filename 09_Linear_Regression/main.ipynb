```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({
    'X':[1400,1600,1700,1875,1100,1550,2350,2450,1425,1700],
    'Y':[245,312,279,300,199,219,405,324,319,255]
})
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data = df, x='X',y ='Y')
```

```python

model = LinearRegression().fit(df[['X']],df[['Y']])

print(f"Intercept (c):{model.intercept_[0]:.2f}")
print(f"Coefficient(m): {model.coef_[0] [0]:.2f}")
```

```python
predicted_house_price = model.predict([[1800]])
predicted_house_price
```

```python
y_all_predicted = model.predict(df[['X']])
print(y_all_predicted)
sns.scatterplot(data=df, x='X', y='Y')
sns.lineplot(x=df[['X']], y=y_all_predicted.flatten(), colors = 'red')
```

```python
from sklearn.metrics import r2_score
#0 to 1
r2 = r2_score(df['Y'], y_all_predicted)
print(f"{r2}")
```

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
mae = mean_absolute_error(df[['Y']], y_all_predicted)
print(f"{mae}")
rae = root_mean_squared_error(df[['Y']],y_all_predicted)
print(f"{rae}")

```
