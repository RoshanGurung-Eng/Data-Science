```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

prices = pd.Series([10, 12, 15, 14, 13], name = 'Price')
# convert 1d into 2d 
# -1 = figure out automatically (give ne as many rows as need  to fit the data)
# 1 = tells NUmpY to create exactly 1 column
data_2d = prices.values.reshape(-1,1)
data_2d
```

```python
#initialize scalerss
std_scaler = StandardScaler()
minmaxscaler = MinMaxScaler()
robustsclaer = RobustScaler()

#normalized_values = minmaxscaler.fit_transform(data_2d); This is used for to see result in one way
# create a dataframe
results = pd.DataFrame({
    'Original':prices,
    'Standardized (Z-score)':std_scaler.fit_transform(data_2d).flatten(),
    'Normalized (0-1)': minmaxscaler.fit_transform(data_2d).flatten(),
    'Robust (IQR)':robustsclaer.fit_transform(data_2d).flatten()
})

print(results.round(2))
```
