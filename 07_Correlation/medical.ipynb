```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("insurance.csv")
d
df = pd.DataFrame(columns = d.bmi)
df
```

```python
columns = "bmi"
```

```python
#Calculate q1 adn q3 

q1 = d[columns].quantile(0.25)
q3 = d[columns].quantile(0.75)

#2: Calculate the Interquartile Range
IQR = q3 -q1

#3: Define the lower and upper bound
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

print(f"Column: {columns}")
print(f"IQR: {IQR:.2f}")
print(f"Lower Bound: {lower_bound:.2f} | Upper Bound: {upper_bound:.2f}")


```
